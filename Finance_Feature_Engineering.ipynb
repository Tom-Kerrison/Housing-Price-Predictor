```python
import pandas as pd

df = pd.read_csv('EPC_Fin.csv')
df.shape
```

```python
int = pd.read_csv('Interest.csv')
inf = pd.read_csv('Inflation.csv')
une = pd.read_csv('Unemployment.csv')
```

```python
une.tail()
```

```python
change = {
    1:'JAN',
    2:'FEB',
    3:'MAR',
    4:'APR',
    5:'MAY',
    6:'JUN',
    7:'JUL',
    8:'AUG',
    9:'SEP',
    10:'OCT',
    11:'NOV',
    12:'DEC',
}

df['MON'] = df['Month'].map(change)
df['Year_Mon'] = df['Year'].astype(str) + ' ' + df['MON']

df['Year_Mon'].head()
```

```python
df.rename(columns={'Year_Mon': 'Title'}, inplace=True)
df.head()
```

```python
mer1 = df.merge(une, on='Title', how='left')
```

```python
mer1['Unemployment rate (aged 16 and over, seasonally adjusted): %'].value_counts()
```

```python
inf.tail()
```

```python
mer2 = mer1.merge(inf, on='Title', how='left')
mer2.head()
```

```python
int.head()
```

```python
df_int = pd.read_csv('Full_Date.csv')
```

```python
df_int.drop(columns=['postcode_clean','match_key','County','Postcode Area','Postcode District'], inplace=True)
df_int.head()
```

```python
df.head()
```

```python
day_df = mer2.merge(df_int, on=['Price (Thousands)', 'Freehold Tenure', 'Leasehold Tenure', 'Old Build',
       'New Build', 'Detached', 'Flat', 'Other Property Type', 'Semi-detached',
       'Terraced', 'Year', 'Month', 'Quarter', 'Day of the Week',
       'Area Code Average Price (Thousands)',
       'Town/City Average Price (Thousands)',
       'District Average Price (Thousands)',
       'County Average Price (Thousands)', 'LAT', 'LONG', 'OSEAST1M',
       'OSNRTH1M'], how="left")
```

```python
day_df.head()
```

```python
day_df[['Year', 'Month', 'Day']] = day_df["Transfer Date"].str.split('-', expand=True)
```

```python
day_df.head()
```

```python
save = day_df
save.shape
```

```python
int.head()
```

```python
day_df['MON'] = day_df['MON'].str.lower()
day_df['MON'] = day_df['MON'].str.capitalize()
```

```python
day_df.drop(columns=['Date'], inplace=True)
```

```python
day_df.rename(columns={'Transfer Date': 'Date'}, inplace=True)
```

```python
day_df.head()
```

```python
day_df['Date Changed'] = pd.to_datetime(
    day_df['Date Changed'],
    format='%d %b %y',
    errors='coerce'  # catches parsing failures as NaT instead of raising
)
int['Date Changed'] = pd.to_datetime(
    int['Date Changed'],
    format='%d %b %y',
    errors='coerce'  # catches parsing failures as NaT instead of raising
)
```

```python
day_df = day_df.sort_values('Date')
int = int.sort_values('Date')
```

```python
day_df1 = day_df.dropna(subset=['Date'])
```

```python
day_df1.shape
```

```python
day_df1['Date'] = pd.to_datetime(day_df1['Date'])
```

```python
mer3 = pd.merge_asof(
    day_df1,
    int,
    on='Date',
    direction='backward'
)
```

```python
mer3.head()
```

```python
mer3.rename(columns={'Unemployment rate (aged 16 and over, seasonally adjusted): %': 'Unemployment rate (%)','CPIH ANNUAL RATE 00: ALL ITEMS 2015=100':'Inflation rate (%)','Rate':'Interest rate (%)'}, inplace=True)
mer3.head()
```

```python
mer3.columns
```

```python
mer3.drop(columns=['MON','Title','Date','Day','Date Changed'])
```

```python
mer3.to_csv('FFE_Final.csv', index=False)
```
